```python
df = pd.read_csv("/content/drive/MyDrive/MLA1_JMA/weather_2012.csv")
```

```python
## Writes the code to display a list of all of the columns in the dataset
df.columns
```

```python
## Renames the column names
df.rename(columns={'Date/Time':'date_time',
                   'Temp (C)':'temp_c',
                   'Dew Point Temp (C)':'dew_point',
                   'Rel Hum (%)':'relative_humidity',
                   'Wind Spd (km/h)':'wind_speed',
                   'Visibility (km)':'visibility',
                   'Stn Press (kPa)':'pressure',
                   'Weather':'weather'
                   },inplace=True)
df.columns
```

```python
## Writes the code to get some standard statistics for each column
df.describe()
```

```python
## Writes the code to add a column named temp_f to the dataset that contains the temperature from temp_c represented in Fahrenheit.
df['temp_f'] = (df['temp_c'] * 9/5) + 32
df.head()
```

```python
## Writes the code to rearrange the columns in the dataset so temp_f appears immediately before temp_c.
df = df[['date_time','temp_f','temp_c','dew_point','relative_humidity','wind_speed','visibility','pressure','weather']]
df.head()
```

```python
## Writes the code to display the data types of all the columns in the dataset
df.dtypes
```

```python
## Writes the code to convert this column to a date/time column and to verify the conversion was successful
df['date_time'] = pd.to_datetime(df['date_time'])
df.dtypes
```

```python
## Writes the code to get the average temperature (in F) for each month
monthly_avg = df.groupby(df['date_time'].dt.month)['temp_f'].mean()
type(monthly_avg)
```

```python
## Writes code to generate a scatterplot that compares the temperature against the relative humidity
import matplotlib.pyplot as plt
plt.scatter(df['temp_f'],df['relative_humidity'])
plt.rcParams['figure.figsize'] = [20,10]
plt.show()
```

```python
## Generates a line chart that shows the average temperature per month
monthly_avg.plot()
plt.rcParams['figure.figsize'] = [20,10]
plt.show()
```

```python
## Writes the code to generate a box plot
df.boxplot()
plt.rcParams['figure.figsize'] = [20,10]
plt.show()
```
